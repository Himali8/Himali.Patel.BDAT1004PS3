# Create a line plot showing the number of marriages and divorces per capita in the U.S. between 1867 and 2014. Label both lines and show the legend.Don't forget to label your axes!


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = pd.read_csv('us-marriages-divorces-1867-2014.csv')
years = a['Year'].values
marriages = a['Marriages_per_1000'].values
divorces = a['Divorces_per_1000'].values
```

```python
plt.figure()
plt.xlim(1867,2014)
plt.plot(years, marriages, label='Marriages')
plt.plot(years, divorces, label='Divorce')
plt.xlabel("Year")
plt.ylabel("Marriage and Divorces")
plt.legend(loc='upper center', ncol=3)
plt.grid(True)
```
